```r
# Split data in train and test
# keep proportion of ham/spam in both datasets
library(caret)

#?createDataPartition
idx = createDataPartition(y = spam$label, times = 1, p = 0.7, list=FALSE )
train = spam[idx, ]
test = spam[-idx, ]
```

```r
prop.table(table(train$label)); prop.table(table(test$label))
```

```r
# cool package to work with text data
library(quanteda)
```

```r
#?tokens
# 1. tokenize text by word
# 2. remove punctuation
# 3. remove numbers
# 4. remove symbols
# 5. remove hyphenation - separate words
# 6. keep html links
```

```r
# some preprocess before using tokens from quanteda package
# some html has spaces inside the address. It is not handled here.

clean_text <- function(text){
    # remove <TEXT> with ' '
    text <- gsub(pattern = '<(.*?)>', replacement = ' ', text)
    
    # replace number with the token "number"
    text <- gsub(pattern = '[0-9]+', replacement = 'number', text)
    
    # replace http and https with the token 'httpaddr
    text <- gsub('(http|https)://[^/s ]*', 'httpaddr', text)
    
    # replace email address with 'emailaddr'
    text <- gsub(pattern = '[a-z0-9^._-]+@[a-z0-9^._-]+', replacement = 'emailaddr', text)
    
    text
    
}
```

```r
train_clean <- clean_text(train$text)
train_clean <- as.data.frame(train_clean)
names(train_clean) <- c('text')
train_clean$label <- train$label
train_clean$text <- as.character(train_clean$text)
head(train_clean, 3)
```

```r
test_clean <- clean_text(test$text)
test_clean <- as.data.frame(test_clean)
names(test_clean) <- c('text')
test_clean$label <- test$label
test_clean$text <- as.character(test_clean$text)
head(test_clean, 3)
```

```r
train_tokens <- tokens(x = train_clean$text, 
                       what = "word",
                       remove_hyphens = TRUE, 
                       remove_punct = TRUE, 
                       remove_symbols = TRUE)
```

```r
test_tokens <- tokens(x = test_clean$text, 
                       what = "word",
                       remove_hyphens = TRUE, 
                       remove_punct = TRUE, 
                       remove_symbols = TRUE)
```

```r
train_tokens <- tokens_tolower(train_tokens)
test_tokens <- tokens_tolower(test_tokens)
```

```r
train_tokens <- tokens_select(train_tokens, stopwords(),
                              selection = "remove",
                              min_nchar = 3L)
test_tokens <- tokens_select(test_tokens, stopwords(),
                             selection = "remove",
                             min_nchar = 3L)
```

```r
train_tokens = tokens_wordstem(train_tokens, language = "english")
test_tokens = tokens_wordstem(test_tokens, language = "english")
```

```r
# Create bag of words only for train_tokens
# We haven't seen the test data. Later we match the test data or new data to the term matrix based on train data
train_tokens_dfm <- dfm(train_tokens, tolower = FALSE)
```

```r
train_tokens_dfm
```

```r
train_tokens_dfm <- dfm_trim(train_tokens_dfm, min_termfreq = 0.001, termfreq_type = "prop")
```

```r
# convert to matrix to check the results
train_tokens_matrix <- as.matrix(train_tokens_dfm)
dim(train_tokens_matrix)
```

```r
# create a data frame with the matrix above
train_tokens_df = cbind(label = train$label, convert(train_tokens_dfm, to = "data.frame"))
head(train_tokens_df)
```

```r
# just in case the column names are in a format not allowed
names(train_tokens_df) <- make.names(names(train_tokens_df))
```

```r
train_tokens_df <- train_tokens_df[,c(-2)]
```

```r
head(train_tokens_df)
```

```r
cv_folds <- createMultiFolds(train$label, k = 10, times = 3)
cv_control <- trainControl(method = "repeatedcv", number = 10,
                         repeats = 3, index = cv_folds)
```

```r
rpart_cv_m1 <- train(label ~ ., data = train_tokens_df, method = "rpart", 
                    trControl = cv_control, tuneLength = 7)
```

```r
rpart_cv_m1
```

```r
confusionMatrix(train_tokens_df$label, predict(object = rpart_cv_m1, newdata = train_tokens_df))
```

```r
plot(rpart_cv_m1$finalModel)
text(rpart_cv_m1$finalModel, cex=.8)
```

```r
# Prepare test set
test_tokens_dfm <- dfm(test_tokens, tolower = FALSE)
```

```r
train_tokens_dfm; test_tokens_dfm
```

```r
# this is the important part.
# The test set or new data coming in it will have different words (new or some of the train data set post-processed)
# You need to match your test/new data to the post-processed train data set
# Now our test/new dataset will have the same columns aka words
# Note: I have seen a lot people THAT create the document-feature matrix using the whole data and then split in train and test set -> WRONG!

test_tokens_dfm <- dfm_select(test_tokens_dfm, pattern = train_tokens_dfm, selection = 'keep')
```
